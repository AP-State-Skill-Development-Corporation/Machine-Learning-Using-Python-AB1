```python
X = df[['sepal length (cm)','petal length (cm)','petal width (cm)']]
y = df['target']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticObject = LogisticRegression()
logisticObject.fit(X,y)
```

```python
logisticObject.score(X,y)
```
